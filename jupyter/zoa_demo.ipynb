# Zoa Jupyter Kernel Demo

This notebook demonstrates the Zoa optical design kernel.
Each cell sends commands to the Zoa engine and displays the output.

**Setup:**
```bash
pip install -e jupyter/
jupyter kernelspec install jupyter/zoa_kernel --user --name=zoa
```

## Load a lens

````zoa
RES osdtriplet
````

## First-order properties

````zoa
FIR
````

## List lens data

````zoa
LIS
````

## Semicolon-separated commands

Multiple commands can be sent in a single cell separated by semicolons.

````zoa
RES osdtriplet; FIR
````

## Paraxial ray trace

````zoa
FIO
````

## Seidel aberrations

````zoa
RSI 0 1
````
